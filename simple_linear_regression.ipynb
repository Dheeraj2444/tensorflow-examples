# Implementation of Linear Regression in TensorFlow
In this notebook, I trained a simple regression line on a small hypothetical 2-D data points to explore tensorflow

### Import required libraries

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

### Make plots within the notebook

```python
%matplotlib inline
```

### Generating random set of 1,000 x,y pairs

```python
data = []
for i in range(1000):
    x = np.random.normal(0.0, 0.6)
    y = 0.1 * x + 0.3 + np.random.normal(0.0, 0.03)
    data.append([x, y])
x_values = [i[0] for i in data]
y_values = [i[1] for i in data]
```

### Visualize input dataset

```python
plt.scatter(x_values, y_values, color='red')
plt.show()
```

### Intializing a weight vector and bias term
Since we have only one independent variable, weight vector would be of shape 1. Generated a random value from Uniform(-1,1)

```python
w = tf.Variable(tf.random_uniform([1], -1, 1))
b = tf.Variable(tf.zeros([1]))
y = w * x_values + b
```

### Loss Function
Using Mean Sqaure Error (MSE) as a loss function and since it is a derivable function we can use the gradient descent for learning purpose 

```python
loss_function = tf.reduce_mean(tf.square(y - y_values))
```

### Model Training
Now that parameters (w and b) are initialized and loss function is defined. We will use a gradient descent learning algorithm in order to find the set of parameters that would minimize the loss function (MSE)


```python
optimizer = tf.train.GradientDescentOptimizer(0.5)  
model = optimizer.minimize(loss_function)
```

### Runing tensorflow session to begin training

```python
init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)
```

### Iterating for 10 epochs; showing regression line along with error and paramters values for each epoch

```python
for i in range(10):
    sess.run(model)
    print ("epoch: ", i, "loss: ", sess.run(loss_function))
    print ("weight: ", sess.run(w), "bias: ", sess.run(b), '\n')
    plt.scatter(x_values, y_values, color='red')
    plt.plot(x_values, sess.run(w) * x_values + sess.run(b))
    plt.show()
```

### Final Output

```python
print ('w = ', sess.run(w), 'b = ', sess.run(b))
```
